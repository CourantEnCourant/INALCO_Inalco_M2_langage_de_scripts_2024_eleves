Le Rain Water Problem (problème de l'eau de pluie) est un problème classique d'algorithmique qui consiste à calculer la quantité d'eau que des barres de différentes hauteurs peuvent accumuler après une pluie.

Énoncé du problème :
On vous donne un tableau d'entiers non négatifs représentant la hauteur de barres verticales sur une grille en 2D, où chaque barre a une largeur de 1 unité. L'objectif est de déterminer combien d'eau (en unités) peut être accumulée entre ces barres après une averse, sachant que l'eau est piégée dans les espaces entre les barres qui ont une hauteur supérieure des deux côtés.

# Solution 1 : comprendre le problème et appliquer

```python
def trap( height):
        """
        :type height: List[int]
        :rtype: int
        """

        def compute_water_blocks_on_level_n(n) :
            rock_index = []
            for i, level in enumerate(height) :
                if level > n :
                    rock_index.append(i)
            return rock_index[-1] +1 - rock_index[0] - len(rock_index)

        total = 0
        max_level = max(height)

        for i in range(max_level) :
            total += compute_water_blocks_on_level_n(i)

        return total
```

Complexité O(max_level×n) Où :

n est la taille de la liste height (nombre de barres).
max_level est la hauteur maximale dans la liste height.

# recherche d'une solution en O(n)

Idée générale : parcours  vertical et pas horizontal
On prend en compte la "géométrie" du problème

```python
def trap( height):

        # On part de chaque côté et on réduit
        left, right = 0, len(height) - 1
        left_max, right_max = height[left], height[right]
        total = 0


        while left < right:
            if left_max < right_max:
                left += 1
                left_max = max(left_max, height[left])
                total += left_max - height[left]

            else:
                right -= 1
                right_max = max(right_max, height[right])
                total += right_max - height[right]

        return total
```
